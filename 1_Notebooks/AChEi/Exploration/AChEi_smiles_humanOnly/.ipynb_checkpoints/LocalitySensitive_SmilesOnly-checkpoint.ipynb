```python
file_path = "Ruack Supplementary material.csv"
df = pd.read_csv(file_path, index_col=0)
```

```python
def to_bit_vector(fp):
    return [int(i) for i in fp.ToBitString()]

from rdkit import Chem
from rdkit.Chem import AllChem
smiles = df['SMILES Notation']
mols = [Chem.MolFromSmiles(smi) for smi in smiles]
bi={}
fps = [AllChem.GetMorganFingerprintAsBitVect(mol, 
                                            radius = 2, 
                                            bitInfo=bi)
      for mol in mols
     ]
fps = np.array([to_bit_vector(fp) for fp in fps])
```

```python
#Remove empty bits
retained_bits=[]
for i in range(fps.shape[1]):
    if np.sum(fps[:, i])>0:
        retained_bits.append(i)
print(f"{len(retained_bits)} out of {fps.shape[1]} bits retained")
```

```python
all_features = fps[:, retained_bits]
all_target = df['Logki'].values
```

```python
import dill as pickle
to_save = {'all_features': all_features, 
           'all_target': all_target, 
           'retained_bits': retained_bits
          }
with open("AChEi_humanOnly_MorganOnly.ob", 'wb') as f:
    pickle.dump(to_save, f)
```

```python
X = tf.cast(all_tensor, dtype=tf.float32)
a_=[]
for i in range(10):
    with tf.GradientTape(persistent=True) as tape:
        tape.watch(X)
        attentions = [model.layers[1].attention_layers[i](X) 
                      for i in range(10)]
        attentions=model.layers[1].concat_layer(attentions)
        with tf.GradientTape() as tape2:
            att0=attentions[:, i*3:(i+1)*3]
            tape2.watch(att0)
            tape2.watch(attentions)
            attentions=tf.concat([attentions[:, :i*3],
                                  att0, attentions[:, (i+1)*3:]],1)
            #Feedforward for remainder of model
            x_weighted = tf.math.multiply(X[:,:n_feat], X[:, n_feat:])
            x_out = tf.einsum('ai, ak->aik', attentions, x_weighted)
            x_out=tf.reshape(x_out, (tf.shape(x_out)[0], 30*n_feat))
            hidden = model.layers[1].dense_layer(x_out)
            eps=np.finfo(np.float32).eps
            act = model.layers[1].activation(hidden+eps)
            act = model.layers[1].batch_norm_layer(act)
            for i in range(2, 8):
                act = model.layers[i](act)
            loss = model.loss(all_target, act)        
        attentions_grad=tape2.gradient(loss, attentions)
    a = tape.gradient(attentions,X)
    a_.append(tape.gradient(att0, X))
```

```python
import dill as pickle
import numpy as np
import pandas as pd
with open("AChEi_humanOnly_Morganonly_attentions_smap.ob", 'rb') as f:
    a_=pickle.load(f)
    
file_path = "Ruack Supplementary material.csv"
df = pd.read_csv(file_path, index_col=0)

import dill as pickle
import numpy as np
with open("AChEi_humanOnly_MorganOnly.ob", 'rb') as f:
    r = pickle.load(f)
    
all_features = r['all_features']
all_target = r['all_target']
retained_bits = r['retained_bits']
all_tensor = np.hstack([all_features, 
                        np.ones_like(all_features).astype(np.float)])

n_feat=len(retained_bits)
```

```python
from rdkit import Chem
from rdkit.Chem import AllChem
from rdkit.Chem.Draw import SimilarityMaps

def update_weights(weights, vbit, val):
    """Helper function for molecule_map"""
    if vbit is None:
        return weights
    for b in vbit:
        weights[b[0]]+=val
        if b[1]>0:
            print(b, val)
    return weights
            
def molecule_map(smi, bit_ind, vbit_val):
    """Maps 'attention' values to atoms in a molecule via Morgan fingerprint bits, thus highlighting atoms of relevant 'high-attention' bits. Currently only highlights single atoms (bits turned on by neighborhoods (radius>0) are printed)
        smi: smiles string of chemical of interest
        bit_ind: (1-dim array) indices of bits (out of 2048bits) of interest (e.g. highlighted by high attention values)
        vbit_val: (1-dim array; same length as bit_ind) attention values for corresponding bit_ind
    """
    
    #Get mol and bit info
    mol=Chem.MolFromSmiles(smi)
    bi={}
    fp = AllChem.GetMorganFingerprintAsBitVect(mol, 
                                               radius=2, 
                                               bitInfo=bi)
    #Filter bit info by relevant attention bits
    vbit_info=[bi.get(b) for b in bit_ind]
    
    #Get atom weights for mapping purposes
    nAtoms=mol.GetNumAtoms()
    weights=[0. for i in range(nAtoms)]
    for i in range(len(vbit_info)):
        weights = update_weights(weights, vbit_info[i], val=vbit_val[i])
    fig = SimilarityMaps.GetSimilarityMapFromWeights(mol, weights)      
    return weights    
```

```python
def get_dl_dx(chem_idx, 
              attentions_grad=attentions_grad, 
              smap_inputs=smap_inputs
             ):
    att0_inv = np.linalg.pinv(np.expand_dims(
        attentions_grad[chem_idx], axis=0
    ))
    da_dx=att0_inv @[smap_inputs[chem_idx]]
    dl_dx = [attentions_grad[chem_idx, i*3:(i+1)*3] @da_dx[i*3:(i+1)*3] 
             for i in range(10)]
    return dl_dx
```

```python
from rdkit import Chem
from rdkit.Chem import AllChem
from rdkit.Chem.Draw import SimilarityMaps

def update_weights(weights, vbit, val):
    """Helper function for molecule_map"""
    if vbit is None:
        return weights
    for b in vbit:
        weights[b[0]]+=val
        if b[1]>0:
            print(b, val)
    return weights
            
def molecule_map(smi, bit_ind, vbit_val):
    """Maps 'attention' values to atoms in a molecule via Morgan fingerprint bits, thus highlighting atoms of relevant 'high-attention' bits. Currently only highlights single atoms (bits turned on by neighborhoods (radius>0) are printed)
        smi: smiles string of chemical of interest
        bit_ind: (1-dim array) indices of bits (out of 2048bits) of interest (e.g. highlighted by high attention values)
        vbit_val: (1-dim array; same length as bit_ind) attention values for corresponding bit_ind
    """
    
    #Get mol and bit info
    mol=Chem.MolFromSmiles(smi)
    bi={}
    fp = AllChem.GetMorganFingerprintAsBitVect(mol, 
                                               radius=2, 
                                               bitInfo=bi)
    #Filter bit info by relevant attention bits
    vbit_info=[bi.get(b) for b in bit_ind]
    
    #Get atom weights for mapping purposes
    nAtoms=mol.GetNumAtoms()
    weights=[0. for i in range(nAtoms)]
    for i in range(len(vbit_info)):
        weights = update_weights(weights, vbit_info[i], val=vbit_val[i])
    fig = SimilarityMaps.GetSimilarityMapFromWeights(mol, weights)      
    return weights    
```
